```python
users =pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/users.csv')
sessions =pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/sessions.csv' )
products =pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/products.csv' )
transactions =pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/transactions.csv')
```

```python
users
```

```python
transactions.head(10)
```

```python
"""12. Join users to transactions, keeping all rows from transactions and only matching
rows from users (left join)"""

pd.merge(transactions,users, how='left', on = "UserID")
```

```python
"""13. Which transactions have a UserID not in users?"""
notinusers=pd.merge(transactions,users, how='left', on = "UserID")
notinusers[notinusers['User'].isnull()][['UserID']]
```

```python
"""14) Join users to transactions, keeping only rows from transactions and users that
match via UserID (inner join)"""

pd.merge(transactions,users, how='inner', on = "UserID")
```

```python
"""15) Join users to transactions, displaying all matching rows AND all non-matching rows
(full outer join)"""
pd.merge(transactions,users, how='outer', on = "UserID")
```

```python
"""16) Determine which sessions occurred on the same day each user registered"""

#sessions.head()
#users.head()

AllUserSession=pd.merge(users,sessions, how='inner', on = ["UserID"])
AllUserSession[AllUserSession["Registered"] == AllUserSession["SessionDate"]]
```

```python
"""17) Build a dataset with every possible (UserID, ProductID) pair (cross join)"""

users['CrossId'] = 1
products['CrossId'] = 1

crossjoin=pd.merge(users,products, how='inner', on = ["CrossId"])
crossjoin[['UserID','ProductID']]
```

```python
"""18) Determine how much quantity of each product was purchased by each user"""

userproducts=pd.merge(users,products, how='inner', on = ["CrossId"])
TotalTransactions=pd.merge(userproducts,transactions, how='left', on = ["UserID","ProductID"])
TotalTransactions[["UserID","ProductID","Quantity"]].fillna(0,axis=1)
```

```python
"""19) For each user, get each possible pair of pair transactions (TransactionID1,
TransacationID2)"""
transactions
users

pd.merge(transactions,transactions, how='left', on = ["UserID"])
```

```python
"""20) Join each user to his/her first occuring transaction in the transactions table"""







mintransactions=transactions.groupby('UserID')[['TransactionDate']].min()

mintransactionssub=pd.merge(transactions,mintransactions, how='inner', on = ["UserID","TransactionDate"])

final=pd.merge(users,mintransactionssub, how='left', on = ["UserID"])
final






```

```python
"""21) Test to see if we can drop columns"""
final.columns
```

```python
## dropped few columns

list(final.dropna(thresh=int(final.shape[0] * .9), axis=1).columns)
```

```python
missing_info=list(final.columns[final.isnull().any()]) 
missing_info
```

```python
for col in missing_info:
    num_missing = final[final[col].isnull() == True].shape[0] 
    print('number missing for column {}: {}'.format(col, num_missing))
```

```python
for col in missing_info:
    percent_missing = final[final[col].isnull() == True].shape[0]/final.shape[0]
    print('percent missing for column {}: {}'.format( col,percent_missing))
```
