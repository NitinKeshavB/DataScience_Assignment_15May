```python
"""1) How-to-count-distance-to-the-previous-zero
For each value, count the difference of the distance from the previous zero (or the start of
the Series, whichever is closer) and if there are no previous zeros, print the position
Consider a DataFrame df where there is an integer column {'X'[7, 2, 0, 3, 4, 2, 5, 0, 3, 4]}
The values should therefore be [1, 2, 0, 1, 2, 3, 4, 0, 1, 2]. Make this a new column 'Y'.
import pandas as pd
df = pd.DataFrame({'X': [7, 2, 0, 3, 4, 2, 5, 0, 3, 4]})"""

import numpy as np
import pandas as pd

df = pd.DataFrame({'X': [7, 2, 0, 3, 4, 2, 5, 0, 3, 4]})
cnt=0 
def DistanceCal(val):
    global cnt
    if val == 0 :
        cnt=0
        return 0
    else:
        cnt+=1
        return cnt
op=df.X.apply(lambda X:DistanceCal(X))
list(op)
```

```python
"""3) Find the sum of the values in s for every Wednesday"""

import datetime
import numpy as np
import pandas as pd
arr=np.random.randint(0,261,261)
Dates=pd.DataFrame(data=arr,columns=['S'])
Dates['DateRange']=pd.date_range(start='2015-01-01', end='2015-12-31', freq='B')
Dates['day_of_week'] = Dates['DateRange'].dt.weekday_name
Dates[Dates['day_of_week'] == 'Wednesday'][['S']].sum()




    
```
